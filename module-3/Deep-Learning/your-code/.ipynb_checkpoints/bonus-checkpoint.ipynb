```python
from math import hypot, cos, sin
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split

"""
A function to generate X/Y data points that will form a spiral.
"""
def spiral(radius, step, resolution=.1, angle=0.0, start=0.0):
    dist = start
    coords=[]
    while dist*hypot(cos(angle),sin(angle))<radius:
        cord=[]
        cord.append(dist*cos(angle))
        cord.append(dist*sin(angle))
        coords.append(cord)
        dist+=step
        angle+=resolution
    return coords
```

```python
# Generate two sets of spiral data points with opposite angles
data_1 = np.array(spiral(15000, 1, angle=0))
data_2 = np.array(spiral(15000, 1, angle=180))
```

```python
# Plot the two datasets to visualize the spirals

import matplotlib.pyplot as plt

a, b = data_1.T
plt.scatter(a, b, s=5)

aa, bb = data_2.T
plt.scatter(aa, bb, s=5)

plt.show()
```

```python
# Combine the two spiral datasets into one

df1 = pd.DataFrame(data=data_1, columns=["X", "Y"])
df1["CLASS"] = 0

df2 = pd.DataFrame(data=data_2, columns=["X", "Y"])
df2["CLASS"] = 1

df = df1.append(df2)
df['CLASS'].value_counts()
```

Now, build a neural network with Tensorflow to classify `df`. See how low data loss and how high accuracy can you achieve!

```python
#populate additional features
df['X**2'] = df['X']**2
df['Y**2'] = df['Y']**2
df['X*Y'] = df['X']*df['Y']
df['sinX'] = df['X'].apply(np.sin)
df['sinY'] = df['Y'].apply(np.sin)
df.head()
```

```python
#split into X and y
X=df.drop(columns=['CLASS']).values
y=df['CLASS'].values

#normalize
X = tf.keras.utils.normalize(X, axis=1)
```

```python
#split into test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
#build model
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(7, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(7, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(7, activation = tf.nn.relu))
model.add(tf.keras.layers.Dense(2, activation = tf.nn.softmax))

optimizer = tf.keras.optimizers.Adam(learning_rate=0.1)
model.compile(
    optimizer=optimizer,
    loss = 'sparse_categorical_crossentropy',
    metrics=['accuracy']
)

X_train.shape, y_train.shape
```

```python
model.fit(X_train, y_train , batch_size=10, epochs = 200)
```
