## Reading track_uri

```python
path = '../dataset/spotify_million_playlist_dataset/data/'
```

```python
t0 = time()
samp = 1
lst = []

filenames = os.listdir(path)

for i, filename in enumerate(sorted(filenames)):
    if filename.startswith("mpd.slice.") and filename.endswith(".json"):
        fullpath = os.sep.join((path, filename))
        f = open(fullpath)
        js = f.read()
        f.close()
        mpd_slice = json.loads(js)
        for playlist in mpd_slice['playlists']:
            for track in playlist['tracks']:
                lst.append([track['track_uri'],playlist['pid']])
    print(filename, i)

    # Time diff
print(f"Time taken: {(time()-t0)/60}")
```

```python
len(lst)
```

```python
lst_uri = [el[0] for el in lst if el[0]]
set_uri = set(lst_uri)
len(set_uri)
```

```python
track_uri = np.array(set_uri)
```

```python
pd.DataFrame(set_uri).to_csv('../data/track_uris.csv', index=None)
```
