### Read playlists - Collaborative Filtering - extract only PID, track_uri and pos 

```python
t0 = time()
keys = [
    'pid',
    'name',
    'description',
    'num_artists',
    'num_albums',
    'num_tracks',
    'num_followers',
    'duration_ms',
    'collaborative',
    'tracks'
]
samp = 2
arr = np.empty(shape=(samp*1000,len(keys)), dtype = object)
filenames = os.listdir(path)
for i, filename in enumerate(random.sample(sorted(filenames), samp)):
    if filename.startswith("mpd.slice.") and filename.endswith(".json"):
        fullpath = os.sep.join((path, filename))
        f = open(fullpath)
        js = f.read()
        f.close()
        mpd_slice = json.loads(js)
        D = pd.DataFrame(mpd_slice['playlists'])[keys].to_numpy()
        arr[i*1000:(i+1)*1000,:]= D
        print(filename,i)
# Time diff
print(f"Time taken: {(time()-t0)/60}")
```
