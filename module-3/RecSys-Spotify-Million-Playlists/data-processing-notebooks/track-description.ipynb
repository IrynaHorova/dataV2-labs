# Extract unique track ids with artis and album features

```python
import json
import numpy as np
from time import time
import pandas as pd
from pandas.io.json import json_normalize
import os
import random
import gc
```

```python
gc.collect()
```

# Reading track_uri

```python
path = '../dataset/spotify_million_playlist_dataset/data/'
path2 = '../data-processed/full-data/track_uris.csv'
```

```python
tracks=pd.read_csv(path2)
tracks_l = tracks['0'].unique()
```

# track_uri, track_name, artist_name, album_name

```python
t0 = time()

T={}
for t in tracks_l:
    T[t] = 0

samp = 1000

filenames = os.listdir(path)
D={}
for i, filename in enumerate(sorted(filenames)[:samp]):
    if filename.startswith("mpd.slice.") and filename.endswith(".json"):
        fullpath = os.sep.join((path, filename))
        f = open(fullpath)
        js = f.read()
        f.close()
        mpd_slice = json.loads(js)
        for playlist in mpd_slice['playlists']:
            for track in playlist['tracks']:
                if T[track['track_uri']] == 0:
                    D[track['track_uri']] =[track['track_name'],track['artist_name'],track["album_name"]]
                    T[track['track_uri']] == 1           
    if i%100==0:
        print(filename, i)

    # Time diff
print(f"Time taken: {(time()-t0)/60}")
```

```python
import json

dict = D

with open('../data-processed/full-data/track_descriptions.json', 'w') as fp:
    json.dump(dict, fp)
```
