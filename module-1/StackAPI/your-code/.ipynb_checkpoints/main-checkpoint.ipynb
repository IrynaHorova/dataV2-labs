# StackAPI

#### Import the necessary libraries here:

```python
# your code here
from stackapi import StackAPI
from datetime import datetime
import pandas as pd
from pandas import json_normalize
```

```python
SITE = StackAPI('stackoverflow')
```

#### Question 1: Find the questions and answers of last month.

```python
# your code here
#fetch questions and answers
questions = SITE.fetch('questions', fromdate = datetime(2020,9,1), todate=datetime(2020,9,30))
answers = SITE.fetch('answers', fromdate=datetime(2020,9,1), todate=datetime(2020,9,30))
```

```python
#flatten questions
nested_columns = ['items']
questions_flat =  json_normalize(questions, nested_columns)
questions_flat.head()
```

```python
#flatten answers
nested_columns = ['items']
answers_flat =  json_normalize(answers, nested_columns)
answers_flat.head()
```

```python
#pd.to_datetime(answers_flat['creation_date'], unit='s')
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
questions_python = SITE.fetch(
    'questions', 
    min = 5, 
    sort = 'votes', 
    fromdate = datetime(2020,10,28), 
    todate = datetime(2020,10,29),
    tagged = 'python'
)
```

```python
# there are no questions with a score of at least 5 today (2020-10-28)
nested_columns = ['items']
questions_python_flat =  json_normalize(questions_python, nested_columns)
questions_python_flat.head()
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
# your code here

answers2 = SITE.fetch(
    'answers', 
    ids = [6784, 6473]
)
```

```python
# there is only one answer from that id list
nested_columns = ['items']
answers2_flat =  json_normalize(answers2, nested_columns)
answers2_flat.head()
```
