```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

import plotly.express as px

%matplotlib inline
```

```python
cdd = pd.read_csv('CDD.csv', sep = '\t')
```

```python
cdd.shape
```

# No 1

#### Print the correlation matrix

```python
#null values
cdd.isnull().sum()
```

```python
#check if Mangnitude if missing for any Geological(erathquake)
cdd[cdd['EVENT TYPE'] == 'Earthquake']['EVACUATED'].isnull().sum()
```

```python
"""I will fill in null values with zeros for the following columns, as they are likely to be zero if NaN:
FATALITIES
INJURED / INFECTED
EVACUATED
MAGNITUDE

governement cost:
FEDERAL DFAA PAYMENTS
PROVINCIAL DFAA PAYMENTS 
PROVINCIAL DEPARTMENT PAYMENTS
MUNICIPAL COSTS

I will not replace other columns as it is not clear to me if they are 0 if not given
"""
```

```python
cdd['FATALITIES'] = cdd['FATALITIES'].fillna(0)
cdd['INJURED / INFECTED'] = cdd['INJURED / INFECTED'].fillna(0)
cdd['EVACUATED'] = cdd['EVACUATED'].fillna(0)
cdd['MAGNITUDE'] = cdd['MAGNITUDE'].fillna(0)
cdd['FEDERAL DFAA PAYMENTS'] = cdd['FEDERAL DFAA PAYMENTS'].fillna(0)
cdd['PROVINCIAL DFAA PAYMENTS'] = cdd['PROVINCIAL DFAA PAYMENTS'].fillna(0)
cdd['PROVINCIAL DEPARTMENT PAYMENTS'] = cdd['PROVINCIAL DEPARTMENT PAYMENTS'].fillna(0)
cdd['MUNICIPAL COSTS'] = cdd['MUNICIPAL COSTS'].fillna(0)
```

```python
#descriptove statistics
cdd.describe().transpose()
```

```python
cdd.corr()
```

#### Find 3 pairs of columns, one pair that is highly positively correlated, one for which the correlation is almost zero and one pair that is negatively correlated. In each case write down an explanation why we see this correlation

```python
""" 
Highly positively correlated:

FEDERAL DFAA PAYMENTS should be highly correlated with ESTIMATED TOTAL COST, because the ESTIMATED TOTAL COST 
is probably driving the pay out
"""

cdd['FEDERAL DFAA PAYMENTS'].corr(cdd['ESTIMATED TOTAL COST'])
```

```python
#positively correlated
plt.scatter(x = cdd['FEDERAL DFAA PAYMENTS'], y = cdd['ESTIMATED TOTAL COST'])
```

```python
""" 
Correlation is almost zero:

MAGNITUDE is recorded only for th earthquakes, so for the vast majority of events it is going to be zero and not
correlate with FATALITIES. You can also see this in the scatterplot below
"""

cdd['FATALITIES'].corr(cdd['MAGNITUDE'])
```

```python
""" 
Correlation is negative:

There are no highly negatively correlated column pairs. 

The highest negative correlation si between 
'UTILITY - PEOPLE AFFECTED (the amount of people whose utility services 
(power, water, etc.) were interrupted/affected by a specific event)
and 'OGD COSTS' (the cost, in dollars, as reported in Open Government Data sources)

I woudl expect positive correlation between these two, so mgiht be the case of data collection,
for example OGD COSTS are available only for 14 out of 1089 records in the dataset, and 
'UTILITY - PEOPLE AFFECTED' are mostly 0, so this explains very weak correlation

"""
cdd['UTILITY - PEOPLE AFFECTED'].corr(cdd['OGD COSTS'])
```
