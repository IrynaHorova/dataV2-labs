#### Plot (x, y) and (x, z).

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here

plt.style.use('ggplot')

fig, [ax1, ax2] = plt.subplots(nrows = 1, ncols = 2, figsize = (9,4))

ax1.plot(x, x**2, lw = 1)
ax1.set_title('QUADRATIC PLOT')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')
ax1.set_xscale('linear')
ax1.set_yscale('linear')

ax2.plot(x, np.exp(x), lw = 1)
ax2.set_title('EXPONENTIAL PLOT')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')
ax2.set_xscale('log')
ax2.set_yscale('log')

fig.suptitle('Figures')

plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
plt.style.use('ggplot')

fig, [ax1, ax2] = plt.subplots(nrows = 1, ncols = 2, figsize = (9,4))

ax1.plot(x, x**2, lw = 1)
ax1.set_title('QUADRATIC PLOT')
ax1.set_xlabel('x label')
ax1.set_ylabel('y label')
ax1.set_xscale('linear')
ax1.set_yscale('linear')
ax1.legend(["linear"])

ax2.plot(x, np.exp(x), lw = 1)
ax2.set_title('EXPONENTIAL PLOT')
ax2.set_xlabel('x label')
ax2.set_ylabel('y label')
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.legend(["quadratic"])

fig.suptitle('Figures')

plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4

* Save your figures in a folder called `figures` in your repo. 

```python
mon = fitbit[['Months_encoded', 'Steps']].groupby('Months_encoded').median().reset_index()
wkday = fitbit[['Work_or_Weekend', 'Steps']].groupby('Work_or_Weekend').median().reset_index()
wkday
```

```python
wkday
```

```python
# # MEDIAN STEPS BY MONTH_ENCODED
# # your code here

# # MEDIAN STEPS BY WORK_OR_WEEKEND
# # your code here
# title = 'median steps by month and workday or weekend'

title = 'Median steps by month and workday or weekend'

plt.style.use('ggplot')

fig, ax = plt.subplots(1, 2, figsize = (12,4))

plt.sca(ax[0])
plt.bar(mon['Months_encoded'], mon['Steps'])
plt.xticks(mon['Months_encoded'],['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.title('Median steps by month')
plt.xlim(0, len(mon['Months_encoded'])+1)

plt.sca(ax[1]) 
plt.bar(wkday['Work_or_Weekend'], wkday['Steps'], width = 0.5)
plt.xticks([0,1], ['Weekend','Workday'])
plt.title('Median steps by weekend and workday')

fig.savefig('figures/'+title + '.png',  bbox_inches = "tight")
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
data = fitbit[['Steps', 'Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']]
data.head()
```

```python
# your code here

title = 'Scatterplots of activity minutes vs steps'

plt.style.use('ggplot')

fig, ax = plt.subplots(nrows = 1, ncols = 3, figsize = (12,4))

for pl in range(3):
    plt.sca(ax[pl]) 
    sns.scatterplot(x = 'Steps', y = data.columns[pl+1], data = fitbit, estimator = np.median)\
    .set_title(f'{data.columns[pl+1]} vs Steps')


fig.savefig('figures/'+title + '.png',  bbox_inches = "tight")
```
