```python
import json
import requests
import pandas as pd
from pandas import json_normalize
```

```python
url = 'https://api.rawg.io/api/games/440682?key=13961a85724f47818f318bebae6b82e1'
try:
    # Request API
    json_data = json.dumps(requests.get(url, headers=headers).text)
    json_data = json.loads(json_data)
    # Only include wa.ted keys
    D = {k:v for k,v in json_data.items() if k in include}

    # Clean up dictionary
    D["platforms"] = []
    for platform in json_data["platforms"]:
        D["platforms"].append(platform["platform"]["name"])

    for key in ("developers", "genres", "publishers", "tags"):
        D[key] = []
        for data in json_data[key]:
            D[key].append(data["name"])

    if json_data["esrb_rating"]:
        D["esrb_rating"] = json_data["esrb_rating"]["name"]

    # Save as JSON file
    name = D["id"]
    with open(f"data/{name}.json","w", encoding="utf-8") as f:
        json.dump(D, f)
except:
    print(f"Error with {url}")
```
