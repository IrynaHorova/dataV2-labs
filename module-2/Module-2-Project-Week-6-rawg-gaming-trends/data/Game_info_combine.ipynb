### This script will run over all files in `./data/game_info/` and write to a csv in `./data/game_info.csv`

```python
import json
import os
from pprint import pprint
import csv
import pandas as pd
pd.set_option("display.max_columns", 50)
pd.set_option("display.max_rows", 20)
```

```python
# Output CSV file
csv_file = open(r"./data/game_info.csv", "w", encoding="utf8")
csv_writer = csv.writer(csv_file, lineterminator="\n")

# Loop through each file
for i, file in enumerate(os.listdir(r"./data/game_info/")):
    with open(fr"data/game_info/{file}", "r", encoding="utf8") as f:
        data = json.load(f)

    # Change all list into string seperated by "||"
    for key, val in data.items():
        if isinstance(val, list) and key!= 'ratings' and key !='metacritic_platforms':
            data[key] = "||".join(v.replace("||", "|") for v in val)
    
    # Split this data into different columns and delete it
    if data["added_by_status"]:
        for status in ("yet", "owned", "beaten", "toplay", "dropped", "playing"):
            data[f"added_status_{status}"] = data["added_by_status"].get(status, 0)
    else:
        for status in ("yet", "owned", "beaten", "toplay", "dropped", "playing"):
            data[f"added_status_{status}"] = 0
    del data["added_by_status"]
    
    for rt in ["exceptional", "recommended", "meh", "skip"]:
        data[f"rating_count_{rt}"] = 0
        data[f"rating_pct_{rt}"] = 0
    
    for rt in ["exceptional", "recommended", "meh", "skip"]:
        for el in data['ratings']:
            if el['title'] == rt:
                data[f"rating_count_{rt}"] = el['count']
                data[f"rating_pct_{rt}"] = el['percent']
    
    # Write to file
    if i == 0: csv_writer.writerow(data.keys())
    csv_writer.writerow(data.values())
    
    # Verbose Notification
    if i and i%10000 == 0: print(f"Finished {i} files")
print("Finished everything")

# Save file
csv_file.close()
```
